### **Tutorial Install ANACONDA**

**Apa Itu Anaconda Navigator?**<br>
Anaconda Navigator adalah antarmuka pengguna grafis (GUI) desktop yang disertakan dalam Anaconda® Distribution yang memungkinkan Anda meluncurkan aplikasi dan mengelola paket, lingkungan, dan saluran conda tanpa menggunakan perintah antarmuka baris perintah (CLI). Navigator dapat mencari paket di Anaconda.org atau di Repositori Anaconda lokal. Tersedia untuk Windows, macOS, dan Linux.
