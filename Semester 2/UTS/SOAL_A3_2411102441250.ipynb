````python
# CODING C
# Lengkapi coding berikut dan jalankan

class Stack:
    def __init__(self):
        self.items = []
        
    def item_masuk(self,item):
        return self.items.append(item)
    
    def cek_isi(self):
        return self.items == []
    
    def item_keluar(self):
        return self.items.pop()

def konversiBiner(decNumber):
    remstack = Stack()  # objek dari kelas Stack

    while decNumber > 0:
        rem = decNumber % 2  # hasil modulo 2 dari nilai decNumber
        remstack.item_masuk(rem)
        decNumber = decNumber // 2  # pembagian bulat (floor division)

    binString = ""
    while not remstack.cek_isi():
        binString = binString + str(remstack.item_keluar())
    return binString
    

konversiBiner(10)

````

<blockquote>
Jelaskan codingan diatas berdasarkan algoritmanya dari awal hingga menghasilkan outputnya!
</blockquote>

Jawaban: 

pada class `Stack` yang memiliki sebuah menthod :
  - `item_masuk` yang memasukkan item ke dalam atribut pada index akhir
  - `cek_isi` melakukan sebuah perbandingan apakah nilai atribut bernilai kosong atau tidak
  - `item_keluar` melakukan pengeluaran nilai items di akhir
  
pada fungsi `konversiBiner(decNumber)`
  - `remstack` adalah sebuah objek dari class Stack
  - dan melakukan perulangan jika `decNumber` bernilai lebih besar dari 0
  - `rem` menampung nilai hasil operasi `decNumber % 2`
  - `remstack.item_masuk(rem)` memasukkan nilai rem ke dalam items pada Stack
  - `decNumber` menampung nilai operasi dari `decNumber // 2`

  - `binString` menampung nilai kosong untuk menampung nilai biner dengan tipe string
  - melakukan perulangan jika items kosong atau tidak dan nilai dibalikkan dengan `not`
  - `binString` menampung operasi binString di tambah dengan nilai remstack yang di keluarkan dengan tipe string
  - dan mengembalikan nilai binstring
