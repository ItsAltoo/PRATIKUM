# **Mengenal sqlite3**

SQLite adalah pustaka bahasa C yang mengimplementasikan mesin basis data SQL yang kecil, cepat, mandiri, andal, dan berfitur lengkap. SQLite adalah mesin basis data yang paling banyak digunakan di dunia.

Untuk membuat database di python kita membutuhkan modul `sqlite3`

pertama kita harus membuat database dan membuka database yang di connect ke sqlite3.

```python
import sqlite3

con = sqlite3.connect('data.db')
cur = con.cursor()
```

fungsi `connect` digunakan untuk membuat database baru jika belum ada, jika sudah ada maka akan membuka database tersebut.

dan `cursor` digunakan untuk mengeksekusi pernyataan sql dan mengambil hasilnya.

<br>

setelah menghubungkan ke database dan membuat cursor,kita bisa membuat table dengan `cur.execute()`.


```python
cur.execute(""" 
create table if not exists barang (
    kode_barang text primary key,
    nama_barang text ,
    harga integer
);
 """)
```

fungsi `execute` digunakan untuk mengeksekusi pernyataan sql.

<br>

untuk mengambil semua data dari table kita bisa menggunakan `cur.fetchall()`

```python
for row in cur.fetchall():
    print(row)
```

fungsi `fetchall` digunakan untuk mengambil semua data dari table yang sudah kita buat sebelumnya.
