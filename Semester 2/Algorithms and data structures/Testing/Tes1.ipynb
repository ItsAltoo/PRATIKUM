apa itu rekursif ?

rekursif adalah sebuah perulangan yang dilakukan tanpa operasi ulangan.ia menggunakan sebuah fungsi yang di panggil di dalam fungsi tersebut

## **Rekursif Faktorial**

```python
import time

def faktorial(n):
    if n <= 0:
        return 1
    return n * faktorial(n-1)

start_time = time.time()  # Mulai pencatatan waktu
result = faktorial(1000)
end_time = time.time()  # Akhiri pencatatan waktu

# print("Faktorial 30:", result)
print(f"Waktu eksekusi: {end_time - start_time:.10f} detik")
```

## **Rekursif Fibonacci**

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

start_time = time.time()  # Mulai pencatatan waktu
result = fibonacci(10) # Menghitung Fibonacci ke-30
end_time = time.time()  # Akhiri pencatatan waktu

print("Fibonacci 10:", result)
print(f"Waktu eksekusi: {end_time - start_time:.10f} detik")
```

## **Rekursif Pangkat**

```python
def pangkat(a,n):
    if n <= 0:
        return 1
    return n * pangkat(a,n-1)

start_time = time.time()  # Mulai pencatatan waktu
result = pangkat(2,10) # Menghitung 2^10
end_time = time.time()  # Akhiri pencatatan waktu

print("Pangkat 2^10:", result)
print(f"Waktu eksekusi: {end_time - start_time:.10f} detik")

#ciri-ciri rekursi:
# 1. Memiliki kondisi dasar (base case) yang menghentikan rekursi.
# 2. Memanggil dirinya sendiri dengan argumen yang lebih kecil atau lebih sederhana.
# 3. Menggunakan tumpukan panggilan (call stack) untuk menyimpan status eksekusi.
# 4. Memiliki potensi untuk menyebabkan tumpukan panggilan yang dalam jika tidak dikelola dengan baik.
# 5. Dapat menyebabkan overhead memori dan waktu eksekusi yang lebih tinggi dibandingkan dengan iterasi.
```

```python
def fib(n):
    if n == 0 or n==1:
        return n
    return fib(n-1)+ fib(n-2)

print(fib(10)) 
```

```python
import turtle
myTurtle = turtle.Turtle() 
myWin = turtle.Screen() # membuat layar

def drawSpiral(myTurtle, lineLen):
    if lineLen > 0: # base case jika panjang garis lebih dari 0
        myTurtle.forward(lineLen) # maju ke depan sesuai panjang garis
        myTurtle.right(90) # belok kanan 90 derajat
        drawSpiral(myTurtle, lineLen - 5) # melakukan rekursif dengan mengurangi panjang garis 5
        
drawSpiral(myTurtle, 100) # memanggil fungsi drawSpiral dengan panjang garis 100
myWin.exitonclick() # menunggu klik untuk menutup layar
```

```python
def tree(branchLen, t):
    if branchLen > 5: # base case jika panjang cabang lebih dari 5
        t.forward(branchLen) # maju ke depan sesuai panjang cabang
        t.right(20) # belok kanan 20 derajat
        tree(branchLen - 15, t) # melakukan rekursif dengan mengurangi panjang cabang 15
        t.left(40) # belok kiri 40 derajat
        tree(branchLen - 15, t) # melakukan rekursif dengan mengurangi panjang cabang 15
        t.right(20) # belok kanan 20 derajat
        t.backward(branchLen) # mundur sesuai panjang cabang
        
def main():
    t = turtle.Turtle() # membuat turtle baru
    myWin = turtle.Screen() # membuat layar
    t.left(90) # belok kiri 90 derajat
    t.up() # angkat pena
    t.backward(100) # mundur 100
    t.down() # turunkan pena
    t.color("blue") # warna biru
    tree(75, t) # memanggil fungsi tree dengan panjang cabang 75
    myWin.exitonclick() # menunggu klik untuk menutup layar
main() # memanggil fungsi main
```

## **Bubble Sort**

mengurutkan nilai yang ada di array dari terkecil hingga terbesar

notasi O(n^2) karena ada 2 perulangan

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        print(f"Outer loop {i}: {arr}")
        for j in range(0, n-i-1):
            print(f"Inner loop {j}: {arr}")
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print(bubble_sort([5, 1, 4, 2, 8]))
```

## **DSA Array**

mencari nilai yang terkecil pada array

```python
my_array = [6,6,3,8,5,7]
nilai_awal = my_array[0]

for i in my_array:
    if i < nilai_awal:
        nilai_awal = i
        
print(nilai_awal)
```

## **Print the increasing pattern**

```python
for i in range(1, 6):
    print('* ' * i)
    if i == 5:
        # Print the decreasing pattern
        for j in range(4, 0,-1):
            print('* ' * j)
```

```python
array = [54,234,23,53,1,24,8]

for i in range(len(array)-1):
    for j in range(len(array)-i-1):
        if array[j] > array[j+1]:
            array[j],array[j+1] = array[j+1],array[j]
            
print(array)
```

```python
for i in range(len(array)-1):
    min_index = i
    for j in range(i+1,len(array)):
        if array[j] < array[min_index]:
            min_index = j
            
    min_value = array.pop(min_index)
    array.insert(i,min_value)
    
print(array)
```
