## **2. Materi Praktikum**  

### **2.1 Konsep Dasar Rekursi**  
Rekursi adalah metode pemecahan masalah di mana solusi bergantung pada solusi instance yang lebih kecil dari masalah yang sama. Fungsi rekursif selalu memiliki:  
- **Kasus dasar (base case)**, yang menghentikan rekursi  
- **Langkah rekursif**, yang memecah masalah dan melakukan pemanggilan diri sendiri  

### **2.2 Implementasi Rekursi di Python**  
Contoh sederhana rekursi:  
```python
def faktorial(n):
    if n == 1:
        return 1
    else:
        return n * faktorial(n-1)
```

### **2.3 Visualisasi Rekursi**  
Rekursi juga sering digunakan untuk menghasilkan visualisasi fraktal seperti **spiral** dan **pohon** menggunakan modul `turtle` di Python.

### **2.4 Kompleksitas Rekursi**  
- Penggunaan rekursi biasanya lebih mudah dibaca, namun dapat menyebabkan overhead memori jika tidak diterapkan dengan benar.
- Kompleksitas waktu dan ruang algoritma rekursif perlu dianalisis sesuai jumlah pemanggilan dan kedalaman rekursi.

### **4.1 Percobaan & Latihan 8.1: Rekursi vs Iterasi**  
<img src="1.jpg" width="950px" alt="Deskripsi Gambar" />

---

````python
# Ketikan kode dan penjelasan disini!

def listsum_A(numList):
    thesum = 0 # sebuah variable yang menampung nilai 0
    for i in numList: # melakukan perulangan hingga angka yang dimasukkan
        thesum += i # melakukan pertambahan nilai i + thesum
    return thesum # mengembalikan nilai thesum

print(listsum_A([1,3,5,7,9])) # 25


def listsum_B(numList): 
    if len(numList) == 1: # sebuah base case yang berhenti jika panjang list 1
        return numList[0] # mengembalikan nilai list yang pertama
    else:
        return numList[0] + listsum_B(numList[1:]) # melakukan penjumlahan rekursif,listsum_B(numList[1:]) adalah list yang diambil dari index 1 hingga akhir

print(listsum_B([1,3,5,7,9])) # 25


````


#### **4.2 Percobaan & Latihan 8.2: Konversi Basis Bilangan Rekursif**  
<img src="2.jpg" width="850px" alt="Deskripsi Gambar" />


````python
# Ketikan kode dan penjelasan disini!
def toStr(n,base):
    convertString = "0123456789ABCDEF" # sebuah variable yang menampung hexadecimal
    if n < base: 
        return convertString[n] # mengembalikan nilai convertString[n]
    else:
        return toStr(n // base,base) + convertString[n % base] # melakukan rekursif dengan membagi n dengan base dan mengembalikan nilai convertString[n%base]

print(toStr(1453,16)) # 5B9

# fungsi toStr adalah sebuah perulangan yang mengubah bilangan desimal menjadi hexadecimal dengan metode rekursif
# nilai n digunakan untuk membagi bilangan desimal yang di masukkan
# nilai base digunakan untuk menentukan basis yang di inginkan
````


#### **4.3 Percobaan & Latihan 8.3: Konversi Basis dengan Stack**  
<img src="3.jpg" width="750px" alt="Deskripsi Gambar" />
---


````python
# Ketikan kode dan penjelasan disini!
from pythonds.basic.stack import Stack

rstack = Stack() # membuat stack baru

def toStr(n,base):
    convertString = "0123456789ABCDEF" # sebuah variable yang menampung hexadecimal
    
    while n > 0: # melakukan perulangan hingga n lebih dari 0
        if n < base:
            rstack.push(convertString[n])
        else:
            rstack.push(convertString[n % base])
        n = n // base
        
    res = "" # sebuah variable yang menampung string kosong
    while not rstack.isEmpty(): # melakukan perulangan hingga stack kosong,ini adalah base case
        res = res + str(rstack.pop()) # mengembalikan nilai stack yang di pop
    return res # mengembalikan nilai res


print(toStr(1453,16)) # 5B9
print(toStr(9420,13)) # 4398
print(toStr(123,2)) # 10110101101 menjadi operasi bilangan biner
````

````python
# Ketikan kode dan penjelasan disini!

import turtle
myTurtle = turtle.Turtle() 
myWin = turtle.Screen() # membuat layar

def drawSpiral(myTurtle, lineLen):
    if lineLen > 0: # base case jika panjang garis lebih dari 0
        myTurtle.forward(lineLen) # maju ke depan sesuai panjang garis
        myTurtle.right(90) # belok kanan 90 derajat
        drawSpiral(myTurtle, lineLen - 5) # melakukan rekursif dengan mengurangi panjang garis 5
        
drawSpiral(myTurtle, 100) # memanggil fungsi drawSpiral dengan panjang garis 100
myWin.exitonclick() # menunggu klik untuk menutup layar


````

````python
import turtle

def tree(branchLen, t):
    if branchLen > 5: # base case jika panjang cabang lebih dari 5
        t.forward(branchLen) # maju ke depan sesuai panjang cabang
        t.right(20) # belok kanan 20 derajat
        tree(branchLen - 15, t) # melakukan rekursif dengan mengurangi panjang cabang 15
        t.left(40) # belok kiri 40 derajat
        tree(branchLen - 15, t) # melakukan rekursif dengan mengurangi panjang cabang 15
        t.right(20) # belok kanan 20 derajat
        t.backward(branchLen) # mundur sesuai panjang cabang
        
def main():
    t = turtle.Turtle() # membuat turtle baru
    myWin = turtle.Screen() # membuat layar
    t.left(90) # belok kiri 90 derajat
    t.up() # angkat pena
    t.backward(100) # mundur 100
    t.down() # turunkan pena
    t.color("blue") # warna biru
    tree(75, t) # memanggil fungsi tree dengan panjang cabang 75
    myWin.exitonclick() # menunggu klik untuk menutup layar
main() # memanggil fungsi main
````


#### **4.5 Percobaan & Latihan 8.5: Segitiga Sierpinski Rekursif**  
<img src="5.jpg" width="850px" alt="Deskripsi Gambar" />
---


````python
# Ketikan kode dan penjelasan disini!
import turtle

def drawTriangle(points,color, myTurtle):
    myTurtle.fillcolor(color) # mengisi warna
    myTurtle.up() # angkat pena
    myTurtle.goto(points[0][0], points[0][1]) # pergi ke titik pertama
    myTurtle.down() # turunkan pena
    myTurtle.begin_fill() # mulai mengisi warna
    myTurtle.goto(points[1][0], points[1][1]) # pergi ke titik kedua
    myTurtle.goto(points[2][0], points[2][1]) # pergi ke titik ketiga
    myTurtle.goto(points[0][0], points[0][1]) # pergi ke titik pertama
    myTurtle.end_fill() # selesai mengisi warna


def getMid(p1, p2):
    return ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2) # mengembalikan titik tengah antara p1 dan p2


def sierpinski(points, degree, myTurtle):
    colormap = ['blue', 'red', 'green', 'white', 'yellow','violet', 'orange'] # warna yang digunakan
    drawTriangle(points, colormap[degree], myTurtle) # menggambar segitiga
    
    if degree > 0: # base case jika derajat lebih dari 0
        sierpinski([points[0], getMid(points[0], points[1]), getMid(points[0], points[2])], degree - 1, myTurtle) # melakukan rekursif dengan mengurangi derajat 1
        sierpinski([points[1], getMid(points[0], points[1]), getMid(points[1], points[2])], degree - 1, myTurtle) # melakukan rekursif dengan mengurangi derajat 1
        sierpinski([points[2], getMid(points[2], points[1]), getMid(points[0], points[2])], degree - 1, myTurtle) # melakukan rekursif dengan mengurangi derajat 1


def main2():
    myTurtle = turtle.Turtle() # membuat turtle baru
    myWin = turtle.Screen() # membuat layar
    myPoints = [[-100, -50], [0, 100], [100, -50]] # titik yang digunakan
    sierpinski(myPoints, 3, myTurtle) # memanggil fungsi sierpinski dengan derajat 3
    myWin.exitonclick() # menunggu klik untuk menutup layar
main2() # memanggil fungsi main
````


#### **4.7.6 Percobaan 8.6: Menara Hanoi**  
**Kode Program:**  
```python
def moveTower(height, fromPole, toPole, withPole):
    if height >= 1:
        moveTower(height-1, fromPole, withPole, toPole)
        moveDisk(fromPole, toPole)
        moveTower(height-1, withPole, toPole, fromPole)

def moveDisk(fp, tp):
    print("Moving disk from", fp, "to", tp)

moveTower(3, "A", "B", "C")
```
**Analisis:**  
Output untuk height=3:
Moving disk from A to B
Moving disk from A to C
Moving disk from B to C
Moving disk from A to B
Moving disk from C to A
Moving disk from C to B
Moving disk from A to B

Kasus dasar terdapat pada kondisi height >= 1 (baris 2). Kompleksitas algoritma ini adalah O(2^n) karena setiap disk memerlukan dua langkah rekursif.
