
### **4.2 Percobaan & Latihan 11.2: Traversal Preorder, Inorder, Postorder**
<img src="2.jpg" width="550px" alt="Deskripsi Gambar" />
---


```python
# Ketikan kode dan penjelasan disini!

def print_preorder(root):
    if root:
        print(root.data,end=' ')
        print_preorder(root.left)
        print_preorder(root.right)
        
        
def print_inorder(root):
    if root:
        print_inorder(root.left)
        print(root.data,end=' ')
        print_inorder(root.right)
        
        
def print_postorder(root):
    if root:
        print_postorder(root.left)
        print_postorder(root.right)
        print(root.data,end=' ')

print("Preorder: ",end='')
print_preorder(root)
print("\ninorder: ",end='')
print_inorder(root)
print("\npostorder: ",end='')
print_postorder(root)
```


### **4.3 Percobaan & Latihan 11.3: Binary Search Tree (BST) â€“ Penyisipan dan Pencarian**
<img src="3.jpg" width="650px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!

class BSTNode:
    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key
        
def insert(root,key):
    if root is None:
        return BSTNode(key)
    if key < root.val:
        root.left = insert(root.left,key)
    else:
        root.right = insert(root.right,key)
    return root

def search(root,key):
    if root is None or root.val == key:
        return root
    if key < root.val:
        return search(root.left,key)
    else:
        return search(root.right,key)

bst = None
data = [50,30,20,40,70,60,80]

for num in data:
    bst = insert(bst,num)
    
node = search(bst,60)
print(f'node ditemukan {node.val if node else 'tidak ditemukan'}')



from matplotlib import pyplot as plt
def plot_bst(node, x=0, y=0, dx=1.5, dy=-1.5, ax=None, parent=None):
    if node is None:
        return
    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 5))
        ax.axis('off')
    ax.text(x, y, str(node.val), ha='center', va='center',
            bbox=dict(facecolor='yellow', boxstyle='circle'))
    if parent:
        ax.plot([parent[0], x], [parent[1], y], 'k-', lw=1)
    if node.left:
        plot_bst(node.left, x - dx, y + dy, dx/1.5, dy, ax, (x, y))
    if node.right:
        plot_bst(node.right, x + dx, y + dy, dx/1.5, dy, ax, (x, y))
    if parent is None:
        plt.show()

plot_bst(bst)
```


### **4.4 Percobaan & Latihan 11.4: Traversal Level Order (Breadth-First Traversal)**
<img src="4.jpg" width="650px" alt="Deskripsi Gambar" />
---


```python
# Ketikan kode dan penjelasan disini!

from collections import deque
def print_level_order(root):
    if root is None:
        return
    queue = deque([root])
    while queue:
        node = queue.popleft()
        print(node.val, end=' ')
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
            
print("level-order traversal: ")
print_level_order(bst)

# traversal level-order mengunjungi node pada setiap level pohon dari kiri ke kanan
# sedangkan traversal lainnya mengunjungi node pada setiap level pohon dari atas ke bawah
```


### **4.5 Percobaan & Latihan 11.5: Menghitung Tinggi (Height) Pohon**
<img src="5.jpg" width="450px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!
def height(node):
    if node is None:
        return 0
    else:
        lHeight = height(node.left)
        rHeight = height(node.right)
        return max(lHeight, rHeight) + 1

print("Tinggi pohon: ", height(bst))
```
