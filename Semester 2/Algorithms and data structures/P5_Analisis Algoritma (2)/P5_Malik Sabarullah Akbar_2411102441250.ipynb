### **4.1 Percobaan & Latihan 5.1 – Operasi pada List**
<img src="1.jpg" width="750px" alt="Deskripsi Gambar" />
---




```python
# Ketikan kode dan penjelasan disini!
from timeit import Timer

def fungsi1():
    l = []
    for i in range(1000):
        l = l + [i]
        
def fungsi2():
    l = []
    for i in range(1000):
        l.append(i)
       
def fungsi3():
    l = [i for i in range(1000)]
    
def fungsi4():
    l = list(range(1000))
    
t1 = Timer("fungsi1()", "from __main__ import fungsi1")
print("concat ", t1.timeit(number=1000), "detik")

t2 = Timer("fungsi2()", "from __main__ import fungsi2")
print("append ", t2.timeit(number=1000), "detik")

t3 = Timer("fungsi3()", "from __main__ import fungsi3")
print("list comprehension ", t3.timeit(number=1000), "detik")

t4 = Timer("fungsi4()", "from __main__ import fungsi4")
print("list constructor ", t4.timeit(number=1000), "detik")

# pada fungsi1() waktu eksekusi paling lama karena menggunakan operator concatenation
# pada fungsi2() waktu eksekusi lebih cepat dari fungsi1() karena menggunakan fungsi append()
# pada fungsi3() waktu eksekusi lebih cepat dari fungsi2() karena menggunakan list comprehension
# pada fungsi4() waktu eksekusi paling cepat karena menggunakan list constructor
```

### **4.2 Percobaan & Latihan 5.2 – Metode `pop()`**
<img src="2.jpg" width="750px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!
x = list(range(1000000))

def pop1():
    x.pop()

def pop2():
    x.pop(0)
    
t1 = Timer("pop1()", "from __main__ import pop1")
print("pop() ", t1.timeit(number=1000), "detik")

t2 = Timer("pop2()", "from __main__ import pop2")
print("pop(0) ", t2.timeit(number=1000), "detik")

# mengapa waktu eksekusi fungsi pop(0) lebih lama dari fungsi pop()?
# karena fungsi pop() hanya menghapus elemen di akhir list, sedangkan fungsi pop(0) menghapus elemen di awal list
# sehingga fungsi pop(0) memerlukan waktu eksekusi yang lebih lama dari fungsi pop()
```


### **4.3 Percobaan & Latihan 5.3 – Pengukuran Waktu `pop()` vs `pop(0)`**
<img src="3.jpg" width="750px" alt="Deskripsi Gambar" />

---

```python
# Ketikan kode dan penjelasan disini!

import timeit
import matplotlib.pyplot as plt


x = list(range(1000000))

pop0_times = []
pop_end_times = []
indices = list(range(1, 21))


for i in indices:
    pop_zero_timer = timeit.Timer("x.pop(0)", "from __main__ import x")
    pop0_time = pop_zero_timer.timeit(number=1000)
    pop0_times.append(pop0_time)
    print([i], "pop(0) ", pop0_time, "detik")

    pop_end_timer = timeit.Timer("x.pop()", "from __main__ import x")
    pop_end_time = pop_end_timer.timeit(number=1000)
    pop_end_times.append(pop_end_time)
    print([i], "pop() ", pop_end_time, "detik")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(indices, pop0_times, marker='o', linestyle='-', color='r', label='pop(0)')
plt.plot(indices, pop_end_times, marker='s', linestyle='-', color='b', label='pop()')
plt.xlabel('Iteration')
plt.ylabel('Time (seconds)')
plt.title('Performance of pop(0) vs pop() in List')
plt.legend()
plt.grid(True)
plt.show()


# pada grafik di atas, waktu eksekusi fungsi pop(0) lebih lama dari fungsi pop() karena fungsi pop(0) menghapus elemen di awal list
# sedangkan fungsi pop() menghapus elemen di akhir list sehingga waktu eksekusi lebih cepat dari fungsi pop(0)
```


### **4.4 Percobaan & Latihan 5.4 – Operasi pada Dictionaries**

```python
import timeit # import modul timeit untuk mengukur waktu eksekusi
import random # import modul random untuk menghasilkan angka acak
import matplotlib.pyplot as plt # import modul matplotlib untuk membuat grafik

list_times = [] # list untuk menyimpan waktu eksekusi fungsi list
dict_times = [] # list untuk menyimpan waktu eksekusi fungsi dictionary
n_values = list(range(10000, 500001, 20000)) # list untuk menyimpan jumlah elemen

for i in n_values: # iterasi sebanyak elemen pada n_values
    t = timeit.Timer("random.randrange(%d) in x" % i, "from __main__ import random, x") # mengukur waktu eksekusi fungsi list
    
    x = list(range(i)) # membuat list dengan jumlah elemen i
    lst_time = t.timeit(number=1000) # mengukur waktu eksekusi fungsi list
    list_times.append(lst_time) # menambahkan waktu eksekusi fungsi list ke dalam list
    
    x = {j: None for j in range(i)} # membuat dictionary dengan jumlah elemen i
    dict_time = t.timeit(number=1000) # mengukur waktu eksekusi fungsi dictionary
    dict_times.append(dict_time) # menambahkan waktu eksekusi fungsi dictionary ke dalam list
    
    print(i, lst_time, dict_time) # menampilkan jumlah elemen, waktu eksekusi fungsi list, dan waktu eksekusi fungsi dictionary

# Plot
plt.figure(figsize=(8, 5))
plt.plot(n_values, list_times, marker='o', linestyle='-', color='r', label='List Lookup')
plt.plot(n_values, dict_times, marker='s', linestyle='-', color='b', label='Dictionary Lookup')
plt.xlabel('Number of Elements')
plt.ylabel('Time (seconds)')
plt.title('Performance of List vs Dictionary Lookup')
plt.legend()
plt.grid(True)
plt.show()
```
