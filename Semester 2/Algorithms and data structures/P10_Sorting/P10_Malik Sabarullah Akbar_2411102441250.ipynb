### **4.1 Percobaan & Latihan 10.1: Bubble Sort**
<img src="1.jpg" width="650px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!

def bubbleSort1(alist):
    # Bubble Sort versi 1: Melakukan perulangan dari akhir ke awal,
    # membandingkan elemen bersebelahan dan menukar jika urutannya salah.
    for passnum in range(len(alist)-1,0,-1):
        for i in range(passnum):
            if alist[i] > alist[i+1]:
                temp = alist[i]
                alist[i] = alist[i+1]
                alist[i+1] = temp
    return alist

def bubbleSort2(alist):
    # Bubble Sort versi 2: Sama seperti versi 1, namun menambahkan flag 'exchanges'
    # untuk menghentikan proses lebih awal jika tidak ada pertukaran pada satu iterasi,
    # artinya data sudah terurut.
    exchanges = True
    passnum = len(alist)-1
    while passnum > 0 and exchanges:
        exchanges = False
        for i in range(passnum):
            if alist[i] > alist[i+1]:
                exchanges = True
                alist[i], alist[i+1] = alist[i+1], alist[i]
        passnum = passnum - 1
    return alist

# pada bubblesort1 lebih mudah di pahami karena menggunakan sedikit variabel

# pada bubblesort1 lebih mudah di pahami karena menggunakan sedikit variabel


myList = [54,26,93,17,77,31,44,55,20]
print(bubbleSort1(myList.copy()))
print(bubbleSort2(myList.copy()))
```


### **4.2 Percobaan & Latihan 10.2: Selection Sort**

```python
# Ketikan kode dan penjelasan disini!

def selectionSort(alist):
    for fillslot in range(len(alist)-1,0,-1): # melakukan perulangan dari belakang
        positionOfMax = 0 # posisi dari nilai terbesar
        for location in range(1,fillslot+1): # melakukan perulangan dari index 1 hingga index 8 + 1
            if alist[location]>alist[positionOfMax]: # membandingkan nilai index 1 dengan index 0
                positionOfMax = location # jika benar lakukan perubahan positionOfMax menjadi location
        temp = alist[fillslot] # menyimpan nilai dari index 8
        alist[fillslot] = alist[positionOfMax] # merubah nilai index 8 menjadi nilai dari positionOfMax
        alist[positionOfMax] = temp # merubah nilai dari positionOfMax menjadi nilai dari index 8
    return alist # mengembalikan nilai alist yang sudah di urutkan

myList = [54,26,93,17,77,31,44,55,20]
print(selectionSort(myList.copy()))
```


### **4.3 Percobaan & Latihan 10.3: Insertion Sort**

<img src="3.jpg" width="650px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!

def insertionSort(alist):
    for index in range(1,len(alist)): # melakukan perulangan dari index 1 hingga index 8
        currentvalue = alist[index] # menyimpan nilai dari index 1
        position = index # menyimpan nilai dari index 1
        while position>0 and alist[position-1]>currentvalue: # jika posisi lebih besar dari 0 dan nilai dari index 0 lebih besar dari currentvalue
            alist[position] = alist[position-1] # merubah nilai dari index 1 menjadi nilai dari index 0
            position = position-1 # mengurangi posisi sebanyak 1
        alist[position] = currentvalue # merubah nilai dari posisi menjadi currentvalue
    return alist # mengembalikan nilai alist yang sudah di urutkan

print(insertionSort(myList.copy()))
```

```python
# Ketikan kode dan penjelasan disini!

def shellSort(alist):
    sublistcount = len(alist)//2 # menghitung panjang dari alist dibagi 2
    while sublistcount > 0: # jika sublistcount lebih besar dari 0
        for startposition in range(sublistcount): # melakukan perulangan dari index 0 hingga index 4
            gapInsertionSort(alist,startposition,sublistcount) # memanggil fungsi gapInsertionSort
        sublistcount = sublistcount // 2 # membagi sublistcount dengan 2
    return alist # mengembalikan nilai alist yang sudah di urutkan

def gapInsertionSort(alist,start,gap):
    for i in range(start+gap,len(alist),gap): # melakukan perulangan dari index 2 hingga index 8 dengan interval gap
        currentvalue = alist[i] # menyimpan nilai dari index 2
        position = i # menyimpan nilai dari index 2
        while position >= gap and alist[position-gap] > currentvalue: # jika posisi lebih besar dari gap dan nilai dari index 0 lebih besar dari currentvalue
            alist[position] = alist[position-gap] # merubah nilai dari index 2 menjadi nilai dari index 0
            position = position - gap # mengurangi posisi sebanyak gap
        alist[position] = currentvalue # merubah nilai dari posisi menjadi currentvalue

print(shellSort(myList.copy()))
```
