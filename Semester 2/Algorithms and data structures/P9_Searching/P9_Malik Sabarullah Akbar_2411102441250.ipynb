### **2.1 Linear Search**  
- **Konsep**: Memeriksa setiap elemen secara berurutan hingga data ditemukan atau seluruh data terproses.  
- **Kompleksitas**: O(n) untuk kasus terburuk.  

### **2.2 Binary Search**  
- **Konsep**: Mencari data pada larik terurut dengan membagi interval pencarian menjadi dua bagian.  
- **Kompleksitas**: O(log n).  

### **2.3 Hashing**  
- **Konsep**: Menggunakan fungsi hash untuk memetakan data ke indeks dalam struktur tabel hash.

### **4.1 Percobaan 9.1: Linear Search**  
<img src="1.jpg" width="550px" alt="Deskripsi Gambar" />

---



```python
# Ketikan kode dan penjelasan disini!

def linear_search(arr,target):
    count = 0 # untuk menampung nilai yang digunakan untuk mengetahui berapa kali perulangan tersebut dilakukan
    for i in range(len(arr)): # melakukan perulangan sebanyak panjang array
        count += 1 # menambah nilai count setiap melakukan perulagan
        if arr[i] == target :  # memandingkan nilai array ke i dengan nilai yang ingin di cari
            return i,count # mengembalikan nilai i dan nilai count
        
    return 'undefind'

data = [12,34,54,2,32]
print(linear_search(data,32))
```


### **4.2 Percobaan 9.2: Binary Search**  
<img src="2.jpg" width="650px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!
def binary_search(arr,target):
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high) //2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

data = [2,12,32,34,54]
print(binary_search(data,54))

data = [24,11,10,24,41,250]
print(binary_search(data,10))
```


### **4.4 Percobaan 9.4: Visualisasi Binary Search**  
<img src="4.jpg" width="650px" alt="Deskripsi Gambar" />
---

```python
# Ketikan kode dan penjelasan disini!
import matplotlib.pyplot as plt
import numpy as np

data = np.sort(np.random.randint(1,100,15))
target = data[7]
steps = []

def binary_search_viz(arr,target):
    low,high = 0,len(arr)-1
    while low <= high:
        mid = (low + high) // 2
        steps.append(mid)
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid -1
    return -1

binary_search_viz(data,target)
plt.plot(data,'ro-',label='Data')
plt.plot(steps,[data[i] for i in steps],'bx--',label='langkah pencarian')
plt.title("Visualisasi binary search")
plt.legend()
plt.show()

# a.step untuk langkah pencarian
# b.karena binary search membagi datanya menjadi dua dan mengambil nilai tengahnya
```
