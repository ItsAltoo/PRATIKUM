### **Soal 1: Perbandingan Fungsi `foo` dan `sumOfN`**

```python
# Ketikan kode dan penjelasan disini!
def foo(n):
    barney = 0
    for bill in range(1,n+1):
        barney = barney + bill
    return barney

def sumOfN(n):
    theSum = 0
    for i in range(1,n+1):
        theSum = theSum + i
    return theSum

print(foo(10),foo(100),foo(1000))
print(sumOfN(10),sumOfN(100),sumOfN(1000))

# metode diatas adalah metode yang sama, hanya saja penamaan variabel yang berbeda
# pada metode foo, variabel bill diganti dengan i dan barney diganti dengan theSum
# pada metode sumOfN, variabel theSum diganti dengan barney dan i diganti dengan bill

print("~"*50)

def foo(n):
    barney = 0
    for i in range(1,n+1):
        barney+=i
    return barney

def sumOfN(n):
    return n*(n+1)//2

print(foo(10),foo(100),foo(1000))
print(sumOfN(10),sumOfN(100),sumOfN(1000))

# metode diatas adalah metode yang sama, hanya saja penamaan variabel yang berbeda
# di bandingkan dengan metode sebelumnya, metode ini lebih efisien karena hanya menggunakan satu operasi
# perhitungan, yaitu n*(n+1)//2
# sedangkan metode sebelumnya menggunakan dua operasi perhitungan, yaitu barney+=i dan theSum = theSum + i
# sehingga metode kedua lebih efisien
```

### **Soal 2 (Code Snippet) – `sumOfN2`**

<img src="2.jpg" width="650px" alt="Deskripsi Gambar" />


```python
# Ketikan kode dan penjelasan disini!
import time
import matplotlib.pyplot as plt

def sumOfN2(n):
    start = time.time()
    theSum = 0
    for i in range(1, n + 1):
        theSum += i
    end = time.time()
    total_time = end - start
    return theSum, total_time

n_values = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
time_values = [sumOfN2(n)[1] for n in n_values]



# jenis notasi big o yang digunakan adalah O(n)

plt.figure(figsize=(8, 5))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='r', label='Execution Time')
plt.xlabel('Value of N')
plt.ylabel('Time (seconds)')
plt.title('grafik waktu eksekusi')
plt.legend()
plt.grid(True)
plt.show()
```

### **Soal 3 – `sumOfN3`**

<img src="3.jpg" width="550px" alt="Deskripsi Gambar" />

```python
# Ketikan kode dan penjelasan disini!
import time

def sumOfN3(n):
    start = time.time()
    result = (n*(n+1))//2
    end = time.time()
    total_time = end - start
    return result, total_time

def sumOfN2(n):
    start = time.time()
    total = 0
    for i in range(1, n + 1):
        total += i
    end = time.time()
    total_time = end - start
    return total, total_time


result_sumOfN3 = sumOfN3(100000)
result_sumOfN2 = sumOfN2(100000)

print("Result of sumOfN3:", result_sumOfN3, "Type:", type(result_sumOfN3))
print("Result of sumOfN2:", result_sumOfN2, "Type:", type(result_sumOfN2))

#perbandingan antara metode sumOfN3 dan sumOfN2
# metode sumOfN3 lebih efisien dibandingkan dengan metode sumOfN2
# karena metode sumOfN3 hanya menggunakan satu operasi perhitungan, yaitu (n*(n+1))//2
# sedangkan metode sumOfN2 menggunakan dua operasi perhitungan, yaitu total += i dan return total
# sehingga metode sumOfN3 lebih efisien

def fibo(n):
    if n <= 1:
        return n
    return fibo(n-1) + fibo(n-2)

print(fibo(6))
```

### **Soal 5 – Checking Off**  

<img src="5.jpg" width="650px" alt="Deskripsi Gambar" />



```python
# Ketikan kode dan penjelasan disini!
def anagramSolution1(s1, s2):
    aList = list(s2)
    pos1 = 0
    stillOK = True
    
    while pos1 < len(s1) and stillOK:
        pos2 = 0
        found = False
        while pos2 < len(aList) and not found:
            if s1[pos1] == aList[pos2]:
                found = True
            else:
                pos2 = pos2 + 1
        if found:
            aList[pos2] = None
        else:
            stillOK = False
        pos1 = pos1 + 1
    return stillOK

a = [1,2,3]
b = [3,2,1]
print(anagramSolution1(a,b))

# analisis kompleksitas checking off (hint:'in' dan 'remove').
# metode checking off menggunakan dua operasi perhitungan, yaitu 'in' dan 'remove'
# operasi 'in' digunakan untuk mengecek apakah elemen yang dicari ada di dalam list
# operasi 'remove' digunakan untuk menghapus elemen yang sudah ditemukan
# kompleksitas dari operasi 'in' adalah O(n) dan kompleksitas dari operasi 'remove' adalah O(n)
# sehingga kompleksitas dari metode checking off adalah O(n^2)
```

### **Soal 6 – Count & Compare**
<img src="6.jpg" width="550px" alt="Deskripsi Gambar" />

```python
# Ketikan kode dan penjelasan disini!
def anagramSolution4(s1, s2):
    c1 = [0]*26
    c2 = [0]*26
    
    for i in range(len(s1)):
        pos = ord(s1[i]) - ord('a')
        c1[pos] = c1[pos] + 1
        
    for i in range(len(s2)):
        pos = ord(s2[i]) - ord('a')
        c2[pos] = c2[pos] + 1
        
    j = 0
    stillOK = True
    while j < 26 and stillOK:
        if c1[j] == c2[j]:
            j = j + 1
        else:
            stillOK = False
    return stillOK

print(anagramSolution4('apple','pleap'))
print(anagramSolution4('orange','ngerao'))
print(anagramSolution4('apple','durian'))

# Untuk list numerik
def numericAnagramSolution(A, B):
    A_sorted = sorted(A)  # Pastikan listA terurut
    B_sorted = sorted(B)  # Pastikan listB terurut
    return A_sorted == B_sorted

# Contoh penggunaan dengan list angka/numerik
A = [1, 2, 3]
B = [3, 2, 1]
print(numericAnagramSolution(A, B))

# analisis notasi big O dari count dan compare 
# metode count dan compare menggunakan dua operasi perhitungan, yaitu count dan compare
# operasi count digunakan untuk menghitung jumlah kemunculan setiap karakter pada kedua string
# operasi compare digunakan untuk membandingkan jumlah kemunculan setiap karakter pada kedua string
# kompleksitas dari operasi count adalah O(n) dan kompleksitas dari operasi compare adalah O(n)
```

### **Soal 7 – Sort & Compare**

<img src="7.jpg" width="450px" alt="Deskripsi Gambar" />


```python
# Ketikan penjelasan disini!
def anagramSolution2(s1, s2):
    aList1 = list(s1)
    aList2 = list(s2)
    
    aList1.sort()
    aList2.sort()
    
    pos = 0
    matches = True
    
    while pos < len(s1) and matches:
        if aList1[pos] == aList2[pos]:
            pos = pos + 1
        else:
            matches = False
    return matches

as1 = anagramSolution2('abcde','edcba')
print(as1)
print(anagramSolution2('abcde','edcb'))
print(anagramSolution2('abcde','edfb'))

print(anagramSolution2('listen','silent'))

# analisis notasi big O dari metode ini 
# operasi sort memiliki kompleksitas O(n log n)
# operasi compare memiliki kompleksitas O(n)
# sehingga kompleksitas dari metode anagramSolution2 adalah O(n log n)
# karena kompleksitas O(n log n) lebih besar dari O(n), maka kompleksitas dari metode anagramSolution2 adalah O(n log n)
```
