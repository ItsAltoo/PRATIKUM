### **4.1 Implementasi Queue**  
<img src="1.jpg" width="550px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!
class Queue:
    def __init__(self):
        self.items = []
        
    def isEmpty(self):
        return self.items == []
    
    def enqueue(self,item):
        self.items.insert(0,item) # menambahkan elemen ke depan
        
    def dequeue(self):
        return self.items.pop() # menghapus elemen dari belakang
    
    def size(self):
        return len(self.items)
    

q = Queue()
q.enqueue(4) # menambahkan elemen ke depan
q.enqueue('dog') 
q.enqueue(True)
print("size :",q.size())
print(q.items)
print("Empty?",q.isEmpty())
print("Dequeue:",q.dequeue()) # menghapus elemen dari belakang 
print("Size:",q.size())
print(q.items)

# mengapa implementasi diatas kurang efisien untuk antrian besar?
# Karena setiap kali kita menambahkan elemen baru, kita harus memindahkan semua elemen yang sudah ada ke belakang, yang membuat waktu kompleksitasnya O(n).

print("~"*20)

# menggunakan metode collections deque
from collections import deque 
class DequeV1:
    def __init__(self):
        self.items = deque() # menggunakan deque dari collections
        
    def isEmpty(self):
        return len(self.items) == []
    
    def addFront(self,item):
        self.items.appendleft(item) # menambahkan elemen ke depan
        
    def addRear(self,item):
        self.items.append(item) # menambahkan elemen ke belakang
        
    def removeFront(self):
        return self.items.popleft() # menghapus elemen dari depan
    
    def removeRear(self):
        return self.items.pop() # menghapus elemen dari belakang
    
    def size(self):
        return len(self.items)

deq = DequeV1()
deq.addFront(1)
deq.addRear(2)
deq.addFront(3)
deq.addRear(4)
print("Size:",deq.size())
print("Empty?",deq.isEmpty())
print(deq.items)
print("Remove Front:",deq.removeFront())
print("Remove Rear:",deq.removeRear())
print("Size:",deq.size())
print(deq.items)


```

### **4.2 Simulasi Hot Potato**  
<img src="2.jpg" width="550px" alt="Deskripsi Gambar" />




```python
# Ketikan kode dan penjelasan disini!
def hotPotato(nameList,num): 
    simqueue = Queue()
    for name in nameList: # menambahkan nama ke dalam antrian
        simqueue.enqueue(name) # menambahkan nama ke dalam antrian
        
    print("Queue:",simqueue.items) # menampilkan antrian awal
    print("-"*20)
    
    while simqueue.size() > 1 : # selama antrian lebih dari 1
        for _ in range(num): # mengulangi sebanyak num
            print("Queue:",simqueue.items) # menampilkan antrian saat ini
            simqueue.enqueue(simqueue.dequeue()) # mengeluarkan elemen dari depan dan menambahkannya ke belakang
        simqueue.dequeue() # mengeluarkan elemen yang terpilih
    return "Nilai akhir :",simqueue.dequeue() # mengeluarkan elemen terakhir yang tersisa

print(hotPotato(["Bill","David","Susan","Jane","Kent","Brand"],7))

```

### **4.3 Implementasi Deque**  
<img src="3.jpg" width="350px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!
class DequeV2 :
    def __init__(self):
        self.items = []
        
    def addFront(self,item):
        self.items.append(item) # menambahkan elemen ke belakang
        
    def addRear(self,item):
        self.items.insert(0,item) # menambahkan elemen ke depan
    
    def removeFront(self):
        return self.items.pop() # menghapus elemen dari belakang
    
    def removeRear(self):
        return self.items.pop(0) # menghapus elemen dari depan
    
    def size(self):
        return len(self.items)
    

d = DequeV2()
d.addRear(4)
d.addFront('car')
d.addRear(8.4)
print("size :",d.size()) # menampilkan ukuran deque
print(d.items)
print("remove rear :",d.removeRear()) # menghapus elemen dari depan
print(d.items)
print("remove front :",d.removeFront()) # menghapus elemen dari belakang
print(d.items)
```


### **4.4 Palindrome Checker**  
<img src="4.jpg" width="550px" alt="Deskripsi Gambar" />

---


```python
# Ketikan penjelasan disini!
def palchecker(aString):
    chardeque = DequeV2()
    for ch in aString:
        print(chardeque.items) 
        chardeque.addRear(ch) # menambahkan karakter ke belakang deque
        
    print("~"*20)
    
    stillEqual = True
    while chardeque.size() > 1 and stillEqual: # selama deque lebih dari 1 dan masih sama
        first = chardeque.removeFront() # menghapus elemen dari depan
        last = chardeque.removeRear() # menghapus elemen dari belakang
        print("first:",first,"last:",last) # menampilkan elemen yang dihapus
        if first != last: # jika elemen yang dihapus tidak sama
            stillEqual = False 
    return stillEqual # mengembalikan nilai True jika palindrom, False jika tidak

print(palchecker("radar"))
print(palchecker("python"))
print(palchecker("leoppoel"))
print(palchecker("leop3poel"))
print(palchecker("leple"))
```


### **4.5 Unordered List**  
<img src="5.jpg" width="450px" alt="Deskripsi Gambar" />


```python
# Ketikan penjelasan disini!
class Node: 
    def __init__(self,initdata): 
        self.data = initdata
        self.next = None
        
class UnorderedList:
    def __init__(self):
        self.head = None
        
    def add(self,item):
        temp = Node(item) # membuat node baru dengan data item
        temp.next = self.head # menghubungkan node baru dengan head
        self.head = temp # mengubah head menjadi node baru
        
        
        
    def size(self):
        current = self.head 
        count = 0
        while current != None: # selama current tidak None
            count += 1
            current = current.next # mengubah current menjadi node berikutnya
            
        return count
    
mylist = UnorderedList()
mylist.add(31)
mylist.add(77)
print("Size :",mylist.size())
```


### **4.6 Ordered List**  
<img src="6.jpg" width="450px" alt="Deskripsi Gambar" />

---


```python
# Ketikan penjelasan disini!
class OrderList(UnorderedList):
    def search(self,item):
        current = self.head 
        found = False
        stop = False
        while current and not found and not stop:
            if current.data == item: # jika data pada node sama dengan item
                found = True
            else:
                if current.data > item: # jika data pada node lebih besar dari item
                    found = False
                current = current.next # mengubah current menjadi node berikutnya
        return found # mengembalikan nilai True jika ditemukan, False jika tidak ditemukan

ol= OrderList()
ol.add(10)
ol.add(5)
ol.add(20)
print("Size :",ol.size())  # menampilkan ukuran ordered list
print("Search 10 :",ol.search(10))
print("Search 5 :",ol.search(5))
print("Search 20 :",ol.search(20)) # mencari elemen yang ada
print("Search 15 :",ol.search(15)) # mencari elemen yang tidak ada
```
