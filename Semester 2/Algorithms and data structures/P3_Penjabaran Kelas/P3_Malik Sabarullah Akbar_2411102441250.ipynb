## **1. Pendahuluan**
Pada praktikum ini, kita akan mempelajari konsep **struktur kontrol** dan **exception handling** dalam Python. Struktur kontrol mencakup percabangan (`if`, `elif`, `else`) dan perulangan (`for`, `while`), yang digunakan untuk mengatur alur eksekusi program berdasarkan kondisi tertentu. Selain itu, **exception handling** (`try`, `except`, `finally`) digunakan untuk menangani kesalahan dalam program agar tidak menyebabkan crash.

## **3. Materi**
### **3.1 Percabangan (Conditional Statements)**
Percabangan digunakan untuk mengeksekusi kode berdasarkan kondisi tertentu.

```python
x = 10
if x > 5:
    print("Nilai x lebih besar dari 5")
elif x == 5:
    print("Nilai x adalah 5")
else:
    print("Nilai x kurang dari 5")
```

### **3.2 Perulangan (Loops)**
#### **3.2.1 Perulangan `for`**
Digunakan untuk mengiterasi elemen dalam suatu koleksi atau rentang nilai.

```python
for i in range(5):
    print("Iterasi ke-", i)
```

#### **3.2.2 Perulangan `while`**
Digunakan untuk mengulang kode selama kondisi tertentu bernilai `True`.

```python
x = 0
while x < 5:
    print("Nilai x:", x)
    x += 1
```

### **3.3 Exception Handling**
Digunakan untuk menangani error dalam program agar tidak menyebabkan crash.

```python
try:
    angka = int(input("Masukkan angka: "))
    hasil = 10 / angka
    print("Hasil:", hasil)
except ZeroDivisionError:
    print("Error: Pembagian dengan nol tidak diperbolehkan!")
except ValueError:
    print("Error: Input harus berupa angka!")
finally:
    print("Program selesai.")
```

---

````python
# Jelaskan apa yang anda pahami dari penjelasan di atas

# struktur kontrol adalah struktur yang digunakan untuk mengatur alur program dengan cara memberikan kondisi tertentu
x = 10
if x > 5:
    print("Nilai x lebih besar dari 5")
elif x == 5:
    print("Nilai x adalah 5")
else:
    print("Nilai x kurang dari 5")
    
# looping adalah struktur yang digunakan untuk mengulang program bisa menggunakan for atau while

# for
for i in range(5):
    print("Iterasi ke-", i)

# while
x = 0
while x < 5:
    print("Nilai x:", x)
    x += 1
    

# exception handling adalah struktur yang digunakan untuk mengatasi error yang terjadi pada program
try:
    angka = int(input("Masukkan angka: "))
    hasil = 10 / angka
    print("Hasil:", hasil)
except ZeroDivisionError:
    print("Error: Pembagian dengan nol tidak diperbolehkan!")
except ValueError:
    print("Error: Input harus berupa angka!")
finally:
    print("Program selesai.")
 
````

````python
# Ketikan kode dan penjelasan disini!

# while adalah salah satu metode perulangan yang digunakan untuk mengulang program dengan jumlah perulangan yang tidak diketahui

angka = 1
while angka > 0 :
    angka = int(input("Masukkan angka (negatif untuk berhenti): "))
print("Program selesai")

# pada program diatas,perulangan akan terus berjalan selama angka yang dimasukkan lebih dari 0,jika angka yang dimasukkan negatif maka program akan berhenti
````

### **Soal 4: Exception Handling**

<img src="4.1.jpg" width="950px" alt="Deskripsi Gambar" />

---



````python
# Ketikan kode dan penjelasan disini!

try : 
    angka1 = int(input("Masukkan angka pertama: "))
    angka2 = int(input("Masukkan angka kedua: "))
    hasil = angka1 / angka2
    print("Hasil:", hasil)
except (ValueError, ZeroDivisionError):
    print("Terjadi kesaalahan pada input anda!")
    
# try digunakan untuk mencoba menjalankan program,except digunakan untuk menangkap error yang terjadi pada program,ValueError digunakan untuk menangkap error jika input yang dimasukkan bukan angka,ZeroDivisionError digunakan untuk menangkap error jika terjadi pembagian dengan nol
````


### **Soal 5: Nested If-Else (Penentuan Grade)**
<img src="5.jpg" width="650px" alt="Deskripsi Gambar" />

````python
# Ketikan kode dan penjelasan disini!
score = int(input("Masukkan nilai: "))
if score >= 80:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
elif score >= 50:
    grade = "D"
else:
    grade = "E"
    
print("Grade:", grade)

# variable score digunakan untuk menyimpan nilai yang dimasukkan oleh user,jika user memasukkan input berupa string maka akan terjadi error,karena input harus berupa angka

# untuk pengondisian diatas,grade A diperoleh jika nilai lebih dari sama dengan 80,grade B diperoleh jika nilai lebih dari sama dengan 70,grade C diperoleh jika nilai lebih dari sama dengan 60,grade D diperoleh jika nilai lebih dari sama dengan 50,dan grade E diperoleh jika nilai kurang dari 50
````



### **Soal 6: Exception Handling (Index Out of Range)**
<img src="6.jpg" width="950px" alt="Deskripsi Gambar" />

````python
# Ketikan kode dan penjelasan disini!

names = ["Alice", "Bob", "Charlie"]

try : 
    index = int(input("Masukkan index (0-2): "))
    print("Name:", names[index])
except ValueError:
    print("Index harus berupa angka!")
except IndexError:
    print("Index diluar batas!")
    
# pada variable names yang berisi list,dan pada try digunakan untuk mencoba menjalankan program memasukkan index yang diketik user dan menampilkan nilai pada list names
# jika ValueError terjadi maka akan menampilkan pesan "Index harus berupa angka!"
# jika IndexError terjadi maka akan menampilkan pesan "Index diluar batas!"
````

````python
# Ketikan penjelasan disini!

# Struktur kontrol (Percabangan & Perulangan)

# Percabangan
angka = 3
if angka % 2 == 0:
    print("Angka", angka, "adalah bilangan genap")
elif angka % 2 != 0:
    print("Angka", angka, "adalah bilangan ganjil")
else:
    print("Angka", angka, "bukan bilangan genap atau ganjil")
    
# - pada variable angka yang berisi angka 3 atau ganjil,yang dimana digunakan untuk perbandingan apakah angka tersebut genap atau ganjil

# - jika angka di modulus 2 hasilnya 0 maka angka tersebut adalah bilangan genap,jika angka yang di masukkan angka ganjil,akan menampilkan pesan "Angka 3 adalah bilangan ganjil"

# - jika variable angka yang dimasukkan bukan bertipe integer maka akan menampilkan opsi yang terdapat pada else

print('~'*50)

# Perulangan

# for
for i in range(1, 10+1):
    print(i)

# - pada perulangan for diatas,akan menampilkan angka dari 1 sampai 10,range digunakan untuk menentukan batas perulangan

# - jika range yang dimasukkan bilangan berkoma maka akan terjadi error,jika minus maka akan menampilkan angka negatif contoh:
for i in range(-10, 10+1):
    print(i)
    
    
# while
while True:
    print("Hello, World!")
    break

# - perulangan while dia berbeda dengan for,perulangan while akan terus berjalan selama kondisi yang dimasukkan benar atau True

# - apa yang terjadi jika kondisi while diatas diubah menjadi False? maka program tidak akan berjalan
# - apa yang terjadi jika kondisi while diatas diubah menjadi True dan menghapus break? maka program akan terus berjalan tanpa berhenti

print('~'*50)

# Exception Handling

try:
    angka = int(input("Masukkan angka: "))
    hasil = 10 / angka
    print("Hasil:", hasil)
except ZeroDivisionError:
    print("Error: Pembagian dengan nol tidak diperbolehkan!")
except ValueError:
    print("Error: Input harus berupa angka!")
finally:
    print("Program selesai.")
    
# - pada program diatas,try digunakan untuk mencoba menjalankan program yang isinya memasukkan angka dan melakukan pembagian angka dengan 10
# - jika terjadi pembagian dengan nol maka akan menampilkan pesan "Error: Pembagian dengan nol tidak diperbolehkan!"
# - jika input yang dimasukkan bukan angka maka akan menampilkan pesan "Error: Input harus berupa angka!"
# - finally digunakan untuk menampilkan pesan "Program selesai." setelah program selesai dijalankan

# - apa yang terjadi jika except ZeroDivisionError dan valueError digabung menjadi satu,xontoh except(ZeroDisionError,valueError)?
# - maka isi dari except tersebut harus dijadikan satu 


# Kesimpulan
# sebutkan contoh kasus nyata di mana struktur  kontrol dan exception handling di gunakan: 

# kalkulator sederhana
try:
    angka1 = int(input("Masukkan angka pertama: "))
    angka2 = int(input("Masukkan angka kedua: "))
    operator = input("Masukkan operator (+, -, *, /): ")
    if operator == '+':
        hasil = angka1 + angka2
    elif operator == '-':
        hasil = angka1 - angka2
    elif operator == '*':
        hasil = angka1 * angka2
    elif operator == '/':
        hasil = angka1 / angka2
    else:
        raise ValueError("Operator yang dimasukkan tidak valid!")
    
    print("Hasil:", hasil)
except ZeroDivisionError:
    print("Error: Pembagian dengan nol tidak diperbolehkan!")
except ValueError as e:
    print("Error:", e)

# - pada program diatas,try digunakan untuk mencoba menjalankan program yang isinya memasukkan angka pertama,angka kedua dan operator yang digunakan
# - jika operator yang dimasukkan tidak sesuai dengan yang diminta maka akan menampilkan pesan "Operator yang dimasukkan tidak valid!" 
# - jika terjadi pembagian dengan nol maka akan menampilkan pesan "Error: Pembagian dengan nol tidak diperbolehkan!"
# - jika input yang dimasukkan bukan angka maka akan menampilkan pesan "Error: Input harus berupa angka!"
````
