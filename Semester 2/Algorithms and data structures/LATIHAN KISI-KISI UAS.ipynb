### **1 STRUKTUR DATA**
---
Buatlah sebuah program menggunakan bahasa Python untuk mengimplementasikan struktur data stack. Program harus menyusun data berupa angka 1345879854 ke dalam stack dengan ketentuan sebagai berikut:

1. Setiap dua digit angka dianggap sebagai satu elemen data.

2. Masukkan setiap elemen ke dalam stack satu per satu (dari kiri ke kanan).

3. Setelah semua data masuk ke dalam stack, tampilkan isi stack dari bawah ke atas.

4. tampilkan setiap elemen dari stack satu per satu hingga kosong.

```python
# Ketikan kode dan penjelasan disini!

class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item)-> None:
        self.items.append(item)
        
    def pop(self)-> str:
        if self.is_empty():
            raise IndexError("pop from empty stack")
        else:
            return self.items.pop()
    
    def is_empty(self)-> bool:
        return len(self.items) == 0
    
    def peek(self)-> str:
        if self.is_empty():
            raise IndexError("peek from empty stack")
        else:
            return self.items[-1]
        
    def size(self)-> int:
        return len(self.items)

num = "1345879854"
data_stack = Stack()

print(5*"~","Show From Front Stack Data", 5*"~")
for i in range(len(num)):
    if i % 2 == 0:
        selected = num[i:i+2]
        print(selected)
        data_stack.push(selected)

        
print(5*"~","Show From Behind Data Stack", 5*"~")
for i in range(data_stack.size() - 1, -1, -1):
    print(data_stack.items[i])

    
print(5*"~","Show Pop Data Stack", 5*"~")
for i in range(len(data_stack.items)):
    print(f"Data ke-{i+1} : {data_stack.pop()}")


print(f"Data Is : {"Empty" if data_stack.is_empty() else "Not Empty" }")
```

*Berikan Penjelasan disini*

Saya membuat class `Stack` yang memiliki metode push,pop,is_empty,peek,dan size.
- Metode `push` digunakan untuk menambahkan elemen ke stack.
- Metode `pop` digunakan untuk menghapus elemen teratas dari stack.
- Metode `is_empty` digunakan untuk memeriksa apakah stack kosong.
- Metode `peek` digunakan untuk melihat elemen teratas stack tanpa menghapusnya.
- Metode `size` digunakan untuk mendapatkan jumlah elemen dalam stack.

### **3 Visualisasi Graf Relasi**
---
Buatlah Graf relasi dengan node yang berasal dari data stack yang anda buat sebelumnya dengan edge yang saing terhubung antar node

```python
import networkx as nx
import matplotlib.pyplot as plt

if not data:
    stack_elements = [num[i:i+2] for i in range(0, len(num), 2)]
else:
    stack_elements = data

grp = nx.Graph()
grp.add_nodes_from(stack_elements)
for i in range(len(stack_elements) - 1):
    for j in range(i + 1, len(stack_elements)):
        grp.add_edge(stack_elements[i], stack_elements[j])
        print(f"Add Edge {stack_elements[i]} - {stack_elements[j]}")

plt.figure(figsize=(10, 6))
pos = nx.spring_layout(grp)
nx.draw(grp, pos, with_labels=True, node_size=800, node_color='#212121', font_size=16, font_color='#eaeaea', font_weight='bold', edge_color='red')
plt.title("Graph Representation of Stack Data", fontsize=20)
plt.show()
```
