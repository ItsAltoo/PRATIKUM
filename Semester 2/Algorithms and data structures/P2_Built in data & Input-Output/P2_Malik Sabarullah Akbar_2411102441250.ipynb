# Built-in Data di Python  
Dalam pemrograman, tipe data merupakan elemen fundamental yang digunakan untuk menyimpan dan memanipulasi data. Python memiliki beberapa **tipe data bawaan (built-in data types)** yang dapat digunakan sesuai kebutuhan dalam pengolahan data. Tipe data ini mencakup angka, teks, boolean, dan berbagai struktur data lainnya.

## Tipe Data Numerik  
Python memiliki tiga tipe data numerik utama:  
- **Integer (`int`)**: Digunakan untuk menyimpan bilangan bulat, misalnya `10`, `-3`, `1000`.  
- **Float (`float`)**: Digunakan untuk menyimpan bilangan desimal atau pecahan, misalnya `3.14`, `0.5`, `-2.7`.  
- **Complex (`complex`)**: Digunakan untuk menyimpan bilangan kompleks yang memiliki bagian nyata dan imajiner, misalnya `3 + 4j`.  

Contoh penggunaan:  

```python
a = 10       # Integer
b = 3.14     # Float
c = 2 + 3j   # Complex

print(type(a))  # Output: <class 'int'>
print(type(b))  # Output: <class 'float'>
print(type(c))  # Output: <class 'complex'>


## Tipe Data Boolean
Tipe data Boolean digunakan untuk merepresentasikan nilai kebenaran, yaitu True (benar) dan False (salah). Tipe data ini sering digunakan dalam operasi logika dan perbandingan.

```python
x = True
y = False

print(type(x))  # Output: <class 'bool'>

a = 10
b = 5

print(a > b)  # Output: True
print(a == b) # Output: False

## Tipe Data Collection
Tipe **data collection** dalam Python adalah jenis struktur data yang digunakan untuk menyimpan beberapa nilai dalam satu variabel. Berbeda dengan tipe data primitif seperti `int` dan `float`, tipe data **collection** memungkinkan kita untuk mengelola dan memanipulasi kumpulan data secara lebih fleksibel.  

Python menyediakan **empat tipe data collection utama**, yaitu:  
1. **List (`list`)** â†’ Kumpulan elemen yang **berurutan** dan dapat **diubah**.  
2. **Tuple (`tuple`)** â†’ Kumpulan elemen yang **berurutan** tetapi **tidak dapat diubah**.  
3. **Set (`set`)** â†’ Kumpulan elemen yang **unik** dan **tidak berurutan**.  
4. **Dictionary (`dict`)** â†’ Koleksi pasangan **key-value** yang memungkinkan pencarian data secara cepat.  

---

````python
# Ketikan kode dan penjelasan disini!

mySet = {3,6,"Kuceng",4.5,False} # variabel yang bertype set
print(mySet)
print(len(mySet)) # menampilkan panjang set
print(False in mySet) # menampilkan nilai False yang ada di dalam set
print("Kuceng" in mySet) # menampilkan nilai Kuceng yang ada di dalam set

print("~"*30)

yourSet = {99,3,100,'kuda'} # variabel yang bertype set

print(mySet.union(yourSet)) # menggabungkan set mySet dan yourSet
print(mySet | yourSet) # menggabungkan set mySet dan yourSet

print(mySet.intersection(yourSet)) # menampilkan nilai yang sama dari set mySet dan yourSet
print(mySet & yourSet) # menampilkan nilai yang sama dari set mySet dan yourSet

print(mySet.difference(yourSet)) # menampilkan nilai yang berbeda dari set mySet dan yourSet
print(mySet - yourSet) # menampilkan nilai yang berbeda dari set mySet dan yourSet

print({3,100}.issubset(yourSet)) # menampilkan nilai yang sama dari set mySet dan yourSet
print({3,100} <= yourSet) # menampilkan nilai yang sama dari set mySet dan yourSet

print("~"*30)

mySet.add("Rumah") # menambahkan nilai Rumah pada set mySet
print(mySet)

mySet.remove(4.5) # menghapus nilai 4.5 pada set mySet
print(mySet)

print(mySet.pop()) # menghapus nilai terakhir pada set mySet
print(mySet)

print("~"*30)

mySet.clear() # menghapus semua nilai pada set mySet
print(mySet)
````
