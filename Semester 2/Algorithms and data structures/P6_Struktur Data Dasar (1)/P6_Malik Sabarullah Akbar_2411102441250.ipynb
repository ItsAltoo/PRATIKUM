### **4.1 Latihan 1 â€“ Implementasi Dasar Stack** 
<img src="1.jpg" width="450px" alt="Deskripsi Gambar" />

**Analisis:**

- Jelaskan proses perubahan isi Stack setelah setiap operasi!

```python
# Ketikan kode dan penjelasan disini!
class Stack: # Kelas Stack untuk implementasi stack
    def __init__(self): # Konstruktor 
        self.items = [] # Inisialisasi stack sebagai list kosong
    
    def isEmpty(self): # Metode untuk memeriksa apakah stack kosong
        return self.items == []
    
    def push(self, item): # Metode untuk menambahkan elemen ke stack
        self.items.append(item)
        
    def pop(self): # Metode untuk menghapus elemen teratas atau terakhir dari stack
        return self.items.pop()

    def peek(self): # Metode untuk melihat elemen teratas atau akhir tanpa menghapusnya
        return self.items[len(self.items) - 1]
    
    def size(self): # Metode untuk melihat panjang stack
        return len(self.items)
    
s = Stack() # Membuat objek stack baru
print(s.isEmpty()) # True
s.push(4) # Menambahkan elemen 4 ke stack
s.push('dog') # Menambahkan elemen 'dog' ke stack
print(s.peek()) # dog
s.push(True) # Menambahkan elemen True ke stack
print(s.size()) # 3
s.push(8.4) # Menambahkan elemen 8.4 ke stack
print(s.pop()) # 8.4
print(s.pop()) # meng
print(s.size()) # 2
```

```python
def parChecker(symbolString): 
    s = Stack() # Membuat objek stack baru
    balanced = True # Inisialisasi variabel balanced sebagai True
    index = 0 # Inisialisasi index sebagai 0
    
    while index < len(symbolString) and balanced: 
        symbol = symbolString[index] 
        if symbol == '([{': # Jika simbol adalah tanda kurung buka
            s.push(symbol)
        else:
            if s.isEmpty():
                balanced = False
            else:
                top = s.pop()
                if not matches(top, symbol):
                    balanced = False
        index = index + 1
        
    if balanced and s.isEmpty(): # Jika semua tanda kurung sudah terpasangkan dengan benar
        return True
    else:
        return False
    
def matches(open, close): # Fungsi untuk memeriksa kecocokan tanda kurung
    opens = '([{'
    closes = ')]}'
    return opens.index(open) == closes.index(close) # Mengembalikan True jika indeks tanda kurung buka dan tutup cocok

print(parChecker('{{([][])}()}')) # false karena tidak ada tanda kurung tutup yang sesuai
print(parChecker('[{()]')) # false karena tidak ada tanda kurung tutup yang sesuai
```


### **4.4 Latihan 4 â€“ Konversi Bilangan Desimal ke Biner** 
<img src="4.jpg" width="550px" alt="Deskripsi Gambar" />



```python
# Ketikan penjelasan disini!
def divideBy2(decNumber): # Fungsi untuk membagi angka desimal dengan 2
    remstack = Stack() # Membuat objek stack baru
    while decNumber > 0: # Selama angka desimal lebih besar dari 0
        rem = decNumber % 2 # Menghitung sisa bagi
        remstack.push(rem) # Menambahkan sisa ke stack
        decNumber = decNumber // 2 # Pembagian bulat
    
    binString = '' # Inisialisasi binString sebagai string kosong
    while not remstack.isEmpty(): # Selama stack tidak kosong
        binString = binString + str(remstack.pop()) # Menggabungkan hasil ke binString
    
    return binString # Mengembalikan hasil

print(divideBy2(42)) # 101010
print(divideBy2(55)) # 110111

# proses konversi dari desimal ke biner menggunakan stack
# adalah sebagai berikut: 
# 1. Buat stack kosong untuk menyimpan sisa bagi.
# 2. Bagi angka desimal dengan 2 dan simpan sisa bagi ke dalam stack.
# 3. Ulangi langkah 2 hingga angka desimal menjadi 0.
# 4. Ambil sisa bagi dari stack satu per satu dan gabungkan menjadi string biner.
# 5. Kembalikan string biner sebagai hasil konversi.

# Mengapa menggunakan stack?
# Karena stack mengikuti prinsip LIFO (Last In First Out), sehingga sisa bagi yang terakhir dimasukkan ke dalam stack adalah yang pertama kali dikeluarkan. Ini sesuai dengan urutan biner yang dihasilkan dari pembagian desimal dengan 2.
```

### **4.5 Latihan 5 â€“ Konversi ke Berbagai Basis Bilangan** 
<img src="5.jpg" width="550px" alt="Deskripsi Gambar" />
---


```python
# Ketikan penjelasan disini!
def baseConverter(decNumber, base): # Fungsi untuk mengonversi angka desimal ke basis lain
    digits = '0123456789ABCDEF' # Daftar digit untuk basis 16
    remstack = Stack() # Membuat objek stack baru
    
    while decNumber > 0: # Selama angka desimal lebih besar dari 0
        rem = decNumber % base # Menghitung sisa bagi
        remstack.push(rem) # Menambahkan sisa ke stack
        decNumber = decNumber // base # Pembagian bulat
    
    newString = '' # Inisialisasi newString sebagai string kosong
    while not remstack.isEmpty(): # Selama stack tidak kosong
        newString = newString + digits[remstack.pop()] # Menggabungkan hasil ke newString
    
    return newString # Mengembalikan hasil

print(baseConverter(25, 2)) # 11001
print(baseConverter(25, 16)) # 19

# perbedaan dari output di atas adalah:
# 1. Pada output pertama, angka desimal 25 dikonversi ke basis 2 (biner) menghasilkan 11001.
# 2. Pada output kedua, angka desimal 25 dikonversi ke basis 16 (heksadesimal) menghasilkan 19.
```


### **4.6 Latihan 6 â€“ Konversi Infix ke Postfix** 
<img src="6.jpg" width="550px" alt="Deskripsi Gambar" />


```python
def infixToPostfix(infixexpr):
    prec = {}
    prec["*"] = 3
    prec["/"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1
    opStack = Stack()
    postfixList = []
    tokenList = infixexpr.split()

    for token in tokenList:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
            postfixList.append(token)
        elif token == "(":
            opStack.push(token)
        elif token == ")":
            topToken = opStack.pop()
            while topToken != "(":
                postfixList.append(topToken)
                topToken = opStack.pop()

        else:
            while (not opStack.isEmpty()) and \
                (prec[opStack.peek()] >= prec[token]):
                postfixList.append(opStack.pop())
            opStack.push(token)

    while not opStack.isEmpty():
        postfixList.append(opStack.pop())

    return " ".join(postfixList)

print(infixToPostfix("A * B + C * D"))
print(infixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )"))
print(infixToPostfix("( A + B ) * ( C + D )")) 
print(infixToPostfix("( A + B ) * C")) # mengkonversi infix ke postfix
print(infixToPostfix("A + B * C")) # mengkonversi infix ke postfix
```


### **4.7 Latihan 7 â€“ Evaluasi Ekspresi Postfix** 
<img src="7.jpg" width="550px" alt="Deskripsi Gambar" />

---


```python
# Ketikan penjelasan disini! 
def postfixEval(postfixExpr): # mendefinisikan fungsi untuk mengevaluasi ekspresi postfix
    operandStack = Stack() # membuat objek stack untuk menyimpan operand
    tokenList = postfixExpr.split() # memisahkan ekspresi postfix menjadi token-token

    for token in tokenList: # melakukan iterasi pada setiap token
        if token in "0123456789": # jika token adalah angka
            operandStack.push(int(token)) # menambahkan angka ke stack
        else:
            operand2 = operandStack.pop() # mengeluarkan operand kedua dari stack
            operand1 = operandStack.pop() # mengeluarkan operand pertama dari stack
            result = doMath(token, operand1, operand2) # melakukan operasi matematika sesuai dengan token 
            operandStack.push(result)   # menambahkan hasil operasi ke stack            

    return operandStack.pop() # mengeluarkan hasil akhir dari stack

def doMath(op, op1, op2):   # mendefinisikan fungsi untuk melakukan operasi matematika
    if op == "*": # jika operasi adalah perkalian
        return op1 * op2
    elif op == "/": # jika operasi adalah pembagian
        return op1 / op2
    elif op == "+": # jika operasi adalah penjumlahan
        return op1 + op2
    else: # jika operasi adalah pengurangan
        return op1 - op2

print(postfixEval('7 8 + 3 2 + /')) # hasilnya 3.0

# cara kerja evaluasi ekspresi postfix menggunakan stack:
# 1. Pisahkan ekspresi postfix menjadi token-token.
# 2. Iterasi pada setiap token.
# 3. Jika token adalah angka, tambahkan ke stack.
# 4. Jika token adalah operator, keluarkan dua operand dari stack.
# 5. Lakukan operasi matematika sesuai dengan operator dan operand yang diambil dari stack.
# 6. Tambahkan hasil operasi ke stack.
# 7. Setelah semua token dievaluasi, keluarkan hasil akhir dari stack.
# 8. Kembalikan hasil akhir sebagai output.

# bagaimana urutan operasi dilakukan pada ekspresi postfix:
# 1. Operand pertama dan kedua diambil dari stack.
# 2. Operasi matematika dilakukan pada operand-operand tersebut.
# 3. Hasil operasi ditambahkan kembali ke stack.
# 4. Proses ini diulang hingga semua token dievaluasi.
# 5. Hasil akhir adalah nilai yang tersisa di stack.
# 6. Urutan operasi dilakukan berdasarkan urutan token dalam ekspresi postfix.
# 7. Operator yang lebih awal dalam ekspresi akan dievaluasi terlebih dahulu.
```
